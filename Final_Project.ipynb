## Data Source

The data is downloaded from [European Centre for Disease Prevention and Control](https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide)

```python
%matplotlib inline
import pandas
```

## A Small Portion of the Data
This is a small portion of data collected for Ireland from the [European Centre for Disease Prevention and Control](https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide) 

```python
df = pandas.read_excel('s3://ia241-1-spring2022-riley/covid_data (2).xls')
Ireland_data = df.loc[ df['countriesAndTerritories']=='Ireland']

Ireland_data[:10]
```

## How Many Coronavirus Cases Have Been Confirmed as of December 2020 in Ireland?

```python
Ireland_data.sum()['cases']
```

```python
print("The number of Coronavirus cases in Ireland as of December 2020 is 76,185 cases.")
```

```python
sum_cases_per_day = Ireland_data.groupby('month').sum()['cases']
sum_cases_per_day.plot()
```

The month with the most confirmed coronavirus cases in Ireland was October. The chart shows a major increase of cases from August to December which is likely due to seasonal changes increasing the liklihood of illness.

## COVID-19 Deaths in Ireland as of December 2020

```python
Ireland_data.sum()['deaths']
```

```python
print("The number of Coronavirus deaths in Ireland as of December 2020 is 2,124.")
```

```python
sum_death_per_month = Ireland_data.groupby('month').sum()['deaths']
sum_death_per_month.plot.bar()
```

The bar chart shows the most deaths from the Coronovaiurs as of December 2020 in Ireland was April.

## Number of Deaths Related to the Number of Cases in Ireland Over Time

This code depicts the trend of Coronavirus deaths in relation to the trend of cases in Ireland as of December 2020. This was calculated by using the Ireland COVID-19 data and assigning variables to the number of deaths, number of cases, and time in months (x=# of cases, y=# of deaths, and c=months) and then making a scatter plot.

```python
Ireland_data.plot.scatter(x='cases',y='deaths',c='month')
```

This scatter plot shows the number of deaths in relation to the month.The most deaths appear to have occured in the first half of the year. Approximately 2.79% of cases resulted in deaths in Ireland.
